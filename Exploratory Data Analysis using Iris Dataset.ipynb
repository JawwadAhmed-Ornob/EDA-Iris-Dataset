```python
sns.FacetGrid(iris, hue= "species", height=4) \
   .map(sns.hisplot, "petal_length") \
   .add_legend();
plt.show();
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'iris' is your DataFrame containing the data
# Corrected the syntax for specifying the size
sns.FacetGrid(iris, hue="species", height=4) \
   .map(sns.boxplot, "petal_length") \
   .add_legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'iris' is your DataFrame containing the data
# Corrected the syntax for specifying the size
g = sns.FacetGrid(iris, hue="species", height=4)

# Map a histogram for petal_length
g.map(sns.histplot, "petal_length")

# Map a line chart for petal_length
g.map(sns.lineplot, "petal_length", estimator='mean', ci=None)

# Add legend
g.add_legend()

# Show the plot
plt.show()
```
